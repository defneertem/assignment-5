## Neural Nets Implementation & XOR Problem

**Goal:** The goal of this exercise is to implement a small neural network from scratch and train it using backprop.

```python
import numpy as np
import pickle
import matplotlib.pyplot as plt
import seaborn as sns
import abc
from tqdm import tqdm
```

```python
# Load xor toy dataset
with open('xordata.pkl', 'rb') as f:
    data = pickle.load(f)

X_train = data['X_train'] # 800 training data points with 2 features
y_train = data['y_train'] # training binary labels {0,1}

X_test = data['X_test']
y_test = data['y_test']
```

```python
def one_hot_encode(y, n_classes):
    return np.identity(n_classes)[y.astype(int)]

def generate_data(dataset_name, n_samples=100):
    match dataset_name:
        case 'regression: an affine line':
            X = np.random.uniform(-5, 5, size=(n_samples, 1))
            y = 3 * X - 100
        case 'regression: cubic polynomial':
            X = np.random.uniform(-5, 5, size=(n_samples, 1))
            y = X ** 3 - 2 * X + 5
        case 'classification: linearly separable':
            X = np.random.normal(size=(n_samples, 2))
            y = one_hot_encode((X[:, 0] > X[:, 1] + 1).astype(int), 2)
        case 'classification: XOR':
            X = np.random.normal(size=(n_samples, 2))
            y = one_hot_encode((X[:, 0] * X[:, 1] > 0).astype(int), 2)
        case _:
            raise ValueError('invalid dataset name')
    return X, y



def train_and_test(model, loss_fn, optimizer, dataset_name, n_epochs=5000, n_samples=100):
    # train
    X_train, y_train = generate_data(dataset_name, n_samples=n_samples)
    losses = []
    for _ in tqdm(range(n_epochs)):
        output = model.forward(X_train)
        loss = loss_fn.forward(output, y_train) 
        losses.append(loss)
        model.backward(loss_fn.backward())
        optimizer.step()

    # test and plot
    X_test, y_test = generate_data(dataset_name, n_samples=n_samples)
    fig, ax = plt.subplots(ncols=2)
    fig.set_size_inches(10, 5)
    ax[0].set_title('loss during training')
    ax[1].set_title('predictions and truth')
    if dataset_name.startswith('classification'):
        # really we should take a softmax here, but this works too (same answer but less interpretable)
        y_pred = np.argmax(model.forward(X_test), axis=1)
        sns.lineplot(losses, ax=ax[0])
        sns.scatterplot(x=X_test[:, 0], y=X_test[:, 1], hue=np.argmax(y_test, axis=1), ax=ax[1])
        sns.scatterplot(x=X_test[:, 0], y=X_test[:, 1], hue=y_pred, s=100, alpha=0.2, ax=ax[1])
        plt.legend([],[], frameon=False)
    elif dataset_name.startswith('regression'):
        y_pred = model.forward(X_test)
        sns.lineplot(losses, ax=ax[0])
        sns.scatterplot(x=X_test[:, 0], y=y_test[:, 0], ax=ax[1])
        sns.scatterplot(x=X_test[:, 0], y=y_pred[:, 0], s=100, alpha=0.2, ax=ax[1])
        plt.legend([],[], frameon=False)
```

```python

# Abstract Base Class for Modules
class Module(abc.ABC):
    __metaclass__ = abc.ABCMeta

    def __init__(self, name):
        self.name = name
        self.weights = {}
        self.grad_weights = {}

    @abc.abstractmethod
    def forward(self, input):
        pass

    @abc.abstractmethod
    def backward(self, grad_output):
        pass

    def __repr__(self):
        return self.name

class Composition(Module): 
    """
    Combines multiple submodules into a sequential pipeline.

    Attributes:
        submodules (list[Module]): List of submodules (e.g., Linear, ReLU) that form the network.
        name (str): String representation of the composition, including names of all submodules.
        weights (dict): Dictionary containing weights and biases from all submodules.

    Methods:
        forward(input): Passes input sequentially through all submodules.
        backward(grad_output): Passes gradients backward through all submodules, updating grad_weights.
    """
    
    def __init__(self, submodules):
        self.submodules = submodules
        self.name = f'Composition({", ".join(submodule.name for submodule in submodules)})'
        self.weights = {
            f'{submodule.name}.{weight_name}': weight
            for submodule in self.submodules
            for weight_name, weight in submodule.weights.items()
        }

    def forward(self, input):
        for submodule in self.submodules:
            input = submodule.forward(input)
        self.output = input
        return input
    
    def backward(self, grad_output):
        for submodule in reversed(self.submodules):
            # the value of submodule.backward is the grad_output for the next
            # submodule
            grad_output = submodule.backward(grad_output)
        # collect all the grad_weights together
        self.grad_weights = {
            f'{submodule.name}.{weight_name}': grad_weight
            for submodule in self.submodules
            for weight_name, grad_weight in submodule.grad_weights.items()
        }
        return grad_output
    

# SGD Optimizer
class SGD:
    """
    Implements Stochastic Gradient Descent (SGD) optimization.

    Attributes:
        module (Module): The module (e.g., Composition) containing weights to be optimized.
        lr (float): Learning rate for weight updates.

    Methods:
        step(): Updates weights using the gradients stored in module.grad_weights.
    """
    def __init__(self, module, lr=1e-3):
        self.module = module
        self.lr = lr

    def step(self):
        for weight_name, weight in self.module.weights.items():
            weight -= self.lr * self.module.grad_weights[weight_name]

# MSE Loss
class MSE(Module):
    """
    Implements the Mean Squared Error (MSE) loss function.

    Attributes:
        input_y (np.ndarray): Predicted output from the model.
        input_t (np.ndarray): Ground-truth target values.

    Methods:
        forward(input_y, input_t): Computes the mean squared error loss.
        backward(): Computes the gradient of the loss with respect to the predictions.
    """
    def __init__(self):
        super().__init__('MSE')

    def forward(self, input_y, input_t):
        self.input_y = input_y
        self.input_t = input_t
        return np.mean((input_y - input_t) ** 2)

    def backward(self):
        return 2 * (self.input_y - self.input_t) / len(self.input_t)

# Linear Layer
class Linear(Module):
    """
    A fully connected (dense) layer for neural networks.

    Attributes:
        n_in_features (int): Number of input features.
        n_out_features (int): Number of output features.
        weights (dict): Contains the weight matrix 'W' and bias vector 'b'.
    
    Methods:
        forward(input): Computes the linear transformation Wx + b.
        backward(grad_output): Computes gradients for weights and inputs.
    """

    def __init__(self, n_in_features, n_out_features):
        super().__init__(f'Linear({n_in_features}, {n_out_features})')
        self.n_in_features = n_in_features
        self.n_out_features = n_out_features
        stdv = 1.0 / np.sqrt(self.n_in_features)
        sample_unit_ball = lambda size: np.random.uniform(-stdv, stdv, size)
        self.weights = {
            'W': sample_unit_ball((n_out_features, n_in_features)),
            'b': sample_unit_ball((n_out_features, ))
        }

    def forward(self, input):
        self.input = input # this is our x
        return input @ self.weights['W'].T + self.weights['b']

    def backward(self, grad_output):
        self.grad_weights = {
            'W': grad_output.T @ self.input, # this second term is our gradient
            'b': grad_output.T @ np.ones((len(self.input), )) # this second term is our gradient
        }
        # return the grad_input
        grad_linear = self.weights['W']
        return grad_output @ grad_linear


```

```python
model = Composition([
    Linear(1, 8),
    Linear(8, 1)
])

train_and_test(
    model=model,
    loss_fn=MSE(),
    optimizer=SGD(model, lr=1e-3),
    dataset_name='regression: an affine line',
    n_epochs=100,
    n_samples=100
)
```

```python
class Softmax(Module):
    """
    Implements the Softmax activation function for multi-class classification.

    Attributes:
        n_classes (int): Number of classes in the output.

    Methods:
        forward(input): Computes the softmax probabilities for the input.
        backward(grad_output): Computes the gradient of the loss with respect to the input.
    """

    def __init__(self, n_classes):
        super().__init__(f'Softmax({n_classes})')
        self.n_classes = n_classes

    def forward(self, input):
        p = np.exp(input)
        sums = np.sum(p, axis=1)
        self.output = (p.T / sums.T).T
        return self.output
    
    def backward(self, grad_output):
        # hacks :)
        s = self.output
        diag = np.einsum('ij, jk -> ijk', s, np.identity(s.shape[1]))
        outer = np.einsum('ij, ik -> ijk', s, s)
        grad_softmax = diag - outer
        return np.einsum('ijk, ik -> ij', grad_softmax, grad_output)


class ReLU(Module):
    """
    Implements the ReLU (Rectified Linear Unit) activation function.

    Methods:
        forward(input): Applies the ReLU function to the input.
        backward(grad_output): Computes the gradient of the loss with respect to the input.
    """

    def __init__(self):
        super().__init__('ReLU')

    def forward(self, input):
        self.input = input
        return input * (input > 0)
    
    def backward(self, grad_output):
        grad_relu = (self.input > 0)
        return grad_output * grad_relu


# Softmax with Cross-Entropy Loss Combined
class SoftmaxCrossEntropy(Module):
    """
    Combines Softmax activation with Cross-Entropy loss for multi-class classification.

    Attributes:
        n_classes (int): Number of classes in the output.
        softmax (Softmax): Softmax activation function for probability computation.
        t (np.ndarray): One-hot encoded true labels.
        softmax_output (np.ndarray): Output of the softmax function.

    Methods:
        forward(input, t): Computes the softmax probabilities and cross-entropy loss.
        backward(): Computes the gradient of the loss with respect to the input.
    """

    def __init__(self, n_classes):
        super().__init__(f'CrossEntropy({n_classes})')
        self.n_classes = n_classes
        self.softmax = Softmax(n_classes)

    def forward(self, input, t):
        self.t = t
        self.softmax_output = self.softmax.forward(input)
        # the 1e-6 for numerical stability
        return -np.mean(t * np.log(self.softmax_output + 1e-6))
    
    def backward(self):
        # the goated identity
        return (self.softmax_output - self.t) / len(self.t)

```

```python
model = Composition([
    Linear(1, 8),
    ReLU(),
    Linear(8, 8),
    ReLU(),
    Linear(8, 1),
])

train_and_test(
    model,
    MSE(),
    SGD(model, lr=1e-4),
    'regression: cubic polynomial',
    n_epochs=5000,
    n_samples=100
)
```

```python
model = Composition([
    Linear(2, 8),
    Linear(8, 2),
    # do not apply softmax here!
])

train_and_test(
    model,
    SoftmaxCrossEntropy(2),
    SGD(model, lr=1e-2),
    'classification: linearly separable',
    n_epochs=1000,
    n_samples=100
)
```

```python
H = 32

model = Composition([
    Linear(2, H),
    ReLU(),
    Linear(H, H),
    ReLU(),
    Linear(H, 2)
])

train_and_test(
    model,
    SoftmaxCrossEntropy(2),
    SGD(model, lr=1e-2),
    'classification: XOR',
    n_epochs=5000,
    n_samples=250
)
```

Summary:

In this question, I implemented a neural network from scratch to classify the XOR dataset. The XOR problem, being non-linearly separable, serves as a benchmark for evaluating a neural network's ability to model complex decision boundaries. The architecture consisted of an input layer with 2 nodes, one hidden layer with 32 ReLU-activated units, and an output layer with 2 softmax-activated nodes. Training was performed using the Softmax Cross-Entropy loss function and mini-batch stochastic gradient descent (SGD) optimization.

To support the implementation, I used a modular approach with a well-structured hierarchy of classes. The core of the implementation revolved around the following:


Module (Abstract Base Class): This served as the foundation for all layers and components of the neural network. It provided an interface for the forward and backward methods, ensuring consistency across subclasses.

Composition: This class organized layers into a sequential pipeline. It handled the forward pass by passing inputs through each submodule and aggregated gradients during the backward pass. Additionally, it consolidated the weights and gradients from all layers for efficient management.

Linear (Fully Connected Layer): The Linear class implemented a fully connected layer by maintaining weights and biases. It performed the linear transformation 
Wx+b during the forward pass and computed gradients for both weights and biases during the backward pass.

ReLU (Activation Function): This class introduced non-linearity by applying the ReLU activation function. During the forward pass, it replaced negative values with zero, and during the backward pass, it propagated gradients only for positive inputs.

Softmax and SoftmaxCrossEntropy:

The Softmax class transformed raw logits into probabilities, making the output interpretable for classification tasks.
The SoftmaxCrossEntropy combined softmax with cross-entropy loss, simplifying the computation of gradients for multi-class classification.
MSE (Mean Squared Error Loss): While primarily used for regression tasks, this class computed the mean squared error loss and its gradient.

SGD (Optimizer): The SGD class updated weights and biases based on the computed gradients and a specified learning rate. This iterative optimization process minimized the loss function.

Training involved generating synthetic XOR data using a custom generate_data function, which produced inputs and their corresponding one-hot encoded labels. The train_and_test function trained the model for a specified number of epochs, monitored the loss over time, and visualized the results. The mini-batch implementation allowed for efficient and scalable training.
